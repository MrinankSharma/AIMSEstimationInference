Choosing a mean and covariance function. A mean function with value $3$ seems initially reasonable. As suggested, let's try the RBF kernel, setting the width to be approximately 10 minutes and scale value to be the standard deviation. 

The noise seems to be pretty small looking above. 

```julia
μ(x) = 3;
κ(x₁, x₂) = 0.85 * exp((-1/(2*10)) * LinearAlgebra.norm(x₁ - x₂)^2);
# \sigma_n represents the model noise
σₙ = 0.1;
```

```julia
# K is the full covariance matrix
# x is a vector of training data points
# \kappa is the kernel
function compute_K(x, κ)
    n = size(x)[1]
    K = Array{Float64}(undef, n,n)
    # lazy, for loop
    for i in 1:n
        for j in 1:n
            K[i, j] = κ(x[i], x[j])
        end
    end
    return K
end

function compute_K(x, κ)
    n = size(x)[1]
    K = Array{Float64}(undef, n,n)
    # lazy, for loop
    for i in 1:n
        for j in 1:n
            K[i, j] = κ(x[i], x[j])
        end
    end
    return K
    end;


# x is the matrix of training data points
# x_hat is the predictive points matrix
function compute_k(x̂, x, κ)
    N, d = size(x)
    n, d = size(x̂)
    # k should be a N x n matrix (as the rest is N x 1)
    k = Array{Float64}(undef, N,n)
    for i in 1:N
        for j in 1:n
            k[i, j] = κ(x[i], x̂[j])
        end
    end
    return k
    end;

function compute_predictive_distribution(x̂, X, y, κ, σₙ, σⱼ=0)
    # compute common quantities
    K_train = compute_K(X, κ);
    k_pred_train = compute_k(x̂, x, κ);
    k_pred = compute_K(x̂, κ);
    
    # compute difficult matrix inverse
    mat_inv = LinearAlgebra.inv(K_train + (σₙ^2+σⱼ^2) * I);
    
    mean = LinearAlgebra.transpose(k_pred_train) * mat_inv * y;
    cov = k_pred + σₙ^2 * I - LinearAlgebra.transpose(k_pred_train) * mat_inv * k_pred_train;
end
```
